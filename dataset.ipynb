```python
from matplotlib import pyplot as plt




free_sulfur_dioxide =  list(df['free_sulfur_dioxide'])
total_sulfur_dioxide =  list(df['total_sulfur_dioxide'])

#dados_treino = 
#dados_teste =


print(total_sulfur_dioxide)

print(df['total_sulfur_dioxide'])
plt.scatter(free_sulfur_dioxide, total_sulfur_dioxide)
```

```python
import numpy as np
import random

x = free_sulfur_dioxide
y = total_sulfur_dioxide


x_treino, x_teste = x[:5198], x[5198:]
y_treino, y_teste = y[:5198], y[5198:]

w = random.random()
print(w)

b = random.random()
print(b)

lr = 0.01

erros = []
for epocas in range(2500):

    for i, x in enumerate(x_treino):
        prediction = w*x + b  #
        
        # ------- y = ax + b -------------------

        # prediction  o y-hat.
        # w  igual ao peso que ser atualizado.
        # elem  o numero dentro do array x.
        # b  onde corta o eixo y

        y = y_treino[i]

        erro = prediction - y
    

        #atualizar a partir do erro 

        dw = (1 / len(x_treino)) * x * erro
        db = (1 / len(x_treino)) * erro

        w = w - lr * dw
        b = b - lr * db

    erros.append(erro)

plt.plot(erros)

plt.show()


        


```
